### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
df.info()
```

```python
# dropna means delete all row in which any value is mising

data_withdropna = df.dropna()
data_withdropna
```

```python
# if we give how='all' it delete row(axis=0) or column(axis=1) if whole row or column have missing value


data_withdropna = df.dropna(how='all',axis=0)
data_withdropna
```

```python
# if how='any' then it will delete any row or column if any 1 value is missing

data_withdropna = df.dropna(how='any',axis=1)
data_withdropna
```

```python
# data fill with fillna
# all missing value tab fill with given parameter in fillna method

data = df.copy()
data_withfillna = data.fillna("xyz")
data_withfillna
```

```python
# it will fill missing value tab with age 35 and cabin with not available 

data_withfillna = data.fillna({'Age':35,'Cabin':'Not Available'})
data_withfillna
```

```python
# this not proprer to fill all tab with same value as that major affect mean 
# so we use variable to fill tab

mean_age = data.Age.mean()
print(mean_age)

data_withfillna = data.fillna({'Age':mean_age,'Cabin':'Not Available'})
data_withfillna
```

```python
# in non-numeric column we can fill value by derive mode and then fill value

cabin = data.Cabin.mode()[0][0:3]
data_withfillna = data.fillna({'Age':mean_age,'Cabin':cabin})
data_withfillna
```

```python
# interpolate method fill tab by interprit value with own algorithem and it only works for numeric column

data_interpolate = data.interpolate()
data_interpolate

# in interpolate method there is a parameter that is limit_direction='forward / backward'
# forward means if last value is NaN then it will not change as after last value there is no row
# backward means if first value is NaN then it will not change as before first value there is no row
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
# Data Normalization
pureData = df.copy()
pureData['Age'] = pureData['Age'].interpolate()
pureData
```

```python
# using min=max normalization technique

data_minmax = pureData.copy()
age_mm = data_minmax['Age']
age_min = age_mm.min()
age_max = age_mm.max()
data_minmax['MinMaxAge'] = (age_mm-age_min)/(age_max-age_min)   # formula X' = X-Xmin / Xmax-Xmin  and (newXmax - newXmin) + newXmin
data_minmax[['Age','MinMaxAge']]
```

```python
# using decimal scaling
# formula X' = X/10^j  where j=no of digit of max value

data_ds = pureData.copy()
age_ds = data_ds['Age']
max_age = age_ds.max()
temp = len(str(int(max_age)))
print(max_age)
print(temp)
data_ds['DecimalScaleAge'] = age_ds / (10**temp)
data_ds[['Age','DecimalScaleAge']]
```

```python
# using z-score
# X' = X-mean /std
data_zs = pureData.copy()
age_zs = data_zs['Age']
age_mean = age_zs.mean()
age_std = age_zs.std()
data_zs['ZScoreAge'] = (age_zs - age_mean) / age_std
data_zs[['Age','ZScoreAge']]
```
